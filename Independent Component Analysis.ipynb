<h2 style="color:rgb(0,120,170)">Task 1: Theoretical aspects concerning ICA: cumulants</h2>

In this task we want to discuss certain theoretical basics of ICA in greater detail. An important aspect of ICA is concerned with measuring non-Gaussianity of random variables. Here we discuss one possible approach in this context that uses cumulants of random variables. These objects are in general defined as follows: for a real-valued random variable $X$ we consider the cumulant generating function $K_{X}$ given by $K_{X}(t)=\ln(\text{E}(\textrm{e}^{Xt}))$. The $n$-th order cumulant $\kappa_n(X)$ corresponding to $X$ is then defined as the $n$-th coefficient in the Taylor-series expansion for $K_{X}(t)$ near $0$, i.e. $K_{X}(t)=\sum_{n=1}^\infty \frac{\kappa_n( X) t^n}{n!}$, and thus $\kappa_n(X)=K_{X}^{(n)}(0)$ (the $n$-th derivative of $K_{X}(t)$ evaluated at $0$). We start by showing some important properties of cumulants which make many calculations (especially regarding sums of independent random variables) easier compared to the use of moments. Let us try to get a better understanding about how these objects look explicitly and can be computed in practice by deriving relations between moments and cumulants.

<h3 style="color:rgb(0,120,170)">Task:</h3>

Assume that $X$ is a centered real random variable, i.e. $\text{E}(X)=0$. Use 
$$
\text{E}(\textrm{e}^{tX})=\sum_{n=0}^\infty \text{E}(X^n) \frac{t^n}{n!}
$$
and the Taylor series expansion of the logarithm near $1$, i.e.
$$
\ln(1+t)=\sum_{n=1}^\infty \frac{(-1)^{n+1} t^n}{n}=t-\frac{t^2}{2}+\frac{t^3}{3}\,\pm\ ...
$$
to compute the first four cumulants in terms of the moments. As a result you should obtain $\kappa_1(X)=0$, $\kappa_2(X)=\text{E}(X^2)$, $\kappa_3(X)=\text{E}(X^3)$ and $\kappa_4(X)=\text{E}(X^4)-3(\text{E}(X^2))^2$.

Hint: You are only interested in an expansion up to order $4$, so you can drop all the higher order terms in all steps of your calculations.

<h3 style="color:rgb(210,90,80)">Calculation (10 points):</h3>

Let $ M_X(t) = \text{E}(e^{tX}) $

Then, we can rewrite it as $ M_X(t) = E(e^{tX}) = \sum_{n=0}^\infty \text{E}(X^n) \frac{t^n}{n!} = \sum_{n=0}^\infty \mu^{'} \frac{t^n}{n!} $

and then, $\mu^{'}_n = \frac{d^n}{dt^n} \text M_X(0) $

So, $ ln(M(t))  =  $ $ln(1+[\mu_1t + \mu_2^{'} \frac{t^2}{2!} + \mu_3^{'} \frac{t^{3}}{3!}+...])$


And, the power series, 

$ x = \mu t + \mu_{2}^{'} \frac{t^2}{2!} + \mu_{3}^{'} \frac{t^3}{3!} +... $

$ x^{2} = \mu^{2}t^{2} + \mu \mu_2^{'}t^{3} + [2 \frac{\mu_3^{'} \mu}{3!} + \frac{( \mu_2^{'})^{2}}{4}]t^{4} + ...  $

$ x^{3} = \mu^{3}t^{3} + 3 \frac{\mu_2^{'} \mu^{2}t^{4}}{2} +...$

$ x^{4} = \mu^{4}t^{4} + ... $


And then, we add up the terms as the power $t^1 $ occurs only in $x$ with coefficient $\mu$. 

So, $\text K(t) = \mu t + [\mu_2^{'} - u^{2}] \frac{t^{2}}{2} + [ \mu_3^{'} -3\mu \mu_2^{'} + 2\mu^{3}] \frac{t^{3}}{3!} + [ \mu_4^{'} - 4 \mu_3^{'} \mu - 3(\mu_2^{'})^{2}+12 \mu_2^{'} \mu^{2} - 6\mu^{4}] \frac{t^{4}}{4!} ... $

Hence,

$ \kappa_1 = \mu $ 

$ \kappa_2 = \mu_2^{'} - \mu^{2} = \text E(X^2) $

$ \kappa_3 = \mu_3^{'} - 3 \mu\mu_2^{'} +2\mu^{3} = \text E[(X- \mu)^{3}] $

$ \kappa_4 = \mu_4^{'} - 4 \mu_3^{'} \mu - 3( \mu_2^{'})^{2} + 12 \mu_2^{'} \mu^{2} - 6 \mu^{4} = \text E[(X- \mu)]^{4} -3 \text (E(X^{2}))^{2}$

As $ \mu = 0 $, then we could write them as;

$ \kappa_1 = 0 $

$ \kappa_2 = \text E(X^2) $

$ \kappa_3 = \text E[(X)^{3}] $

$ \kappa_4 = \text [E(X)^{4}] -3 \text [E(X^{2})]^{2}$

Here we will derive some fundamental properties of cumulants, that make many calculations regarding sums of independent random variables a lot easier and will also provide an important argument, why cumulants can indeed turn out to be useful to measure non-Gaussianty. To this end, show the following: 

Assume that $X_1$ and $X_2$ are independent random variables. Then for any $n \in \mathbb{N}$:

* $\kappa_n(X_1+X_2)=\kappa_n(X_1)+\kappa_n(X_2)$

* $\kappa_n(\alpha X_1)=\alpha^n \kappa_n(X_1)$ for any $\alpha \in \mathbb{R}$

Hint: use that $\text{E}(\textrm{e}^{X_1 t} \textrm{e}^{X_2 t})=\text{E}(\textrm{e}^{X_1 t})\text{E}(\textrm{e}^{X_2 t})$, which follows from independence. You also might want to recall how the exponential function and the logarithm behave with respect to multiplication and addition.

<h3 style="color:rgb(210,90,80)">Calculation (15 points):</h3>

Since $ X_1 and X_2 $ are independent random variables,

then, $ K_{X_1 + X_2} = ln[ \text E(e^{t(X_1 + X_2)})] $

$ K_{X_1 + X_2} = ln \big[ \text E[e^{tX_1}] \text E[e^{tX_2}] \big] $

$ K_{X_1 + X_2} = ln \text E[e^{tX_1}] + ln \text E[e^{tX_2}] $

$ K_{X_1 + X_2} = K_{X_1} (t) + K_{X_2} (t) $

So, it is suffice to show that $ \kappa_n (X_1 + X_2) = \kappa_n(X_1) + \kappa_n(X_2) $

<br/>
<br/>

For t near 0, then we have 

$ K_{aX_1}(t) = K_{X_1}(at) $

$ K^n_{aX_1}(t) = a^nK^n_{X_1}(at) $

$ \kappa_n(aX_{1}) $ = $K^n_{aX_1} (0) $ = $ a^n K^n_{X_1}(0) $ =  $ a^n \kappa_n(X_1) $

<h3 style="color:rgb(0,120,170)">Task:</h3>

From these properties prove the following important observation: let's say we have a standardized (i.e. zero mean and unit variance) sequence of i.i.d. variables $(X_n)_{n=1}^\infty$. Then the rescaled sum $S_n=\frac{X_1+...+X_n}{\sqrt{n}}$ has the property that $\kappa_4(S_n) \to 0$ as $n \to \infty$.

<h3 style="color:rgb(210,90,80)">Calculation (5 points):</h3>

Suppose that $ X_1,..., X_n $ are independent random variables with zero mean as the given conditions, 

We can write it as, $ K_{S_n} = \frac{K_{X_1 + ... + X_n}(t)}{\sqrt{n}} $ 

and then, 
$ K_{\frac{X_1 +...+X_n}{\sqrt{n}}} = K_{X_1} \big (\frac{t}{\sqrt{n}} \big) + ... + K_{X_n} \big( \frac{t}{\sqrt{n}} \big)$

So, we can rewrite then in terms of cumulants, 

$ K_m \Big[\frac{X_1+...+X_n}{\sqrt{n}} \Big] = \frac{K_m(X_1)+...+K_m(X_n)}{n^{m/2}} $

Let suppose that the cumulants are bounded by some constant(C), then 

$ K_m \Big[\frac{X_1+...+X_n}{\sqrt{n}} \Big] \le \frac{nC}{n^{m/2}} $ 

So, when m=4, then $ \frac{nC}{n^2} = \frac{C}{n} $, as $n \to \infty$ and it goes to 0.

Hence, $\kappa_4(S_n) \to 0 $ 



Since $\kappa_4(X)=0$ for a standard normally distributed random variable $X \sim \mathcal{N}(0,1)$, this is in perfect accordance with the central limit theorem, which basically tells you, that sums of normalized and rescaled independent random variables approach $X \sim \mathcal{N}(0,1)$. This should give you a first theoretical explanation why cumulants, and especially the forth cumulant, can be a reasonable concept to measure non-Gaussianity. 

<h3 style="color:rgb(0,120,170)"> Task:</h3>

By doing an explicit calculation, this task of the exercise should give you a further indication how deviation from Gaussianity in terms of cumulants might look by means of a concrete example. To this end, compute the cumulants for a $\lambda$-exponentially distributed variable $X$. Recall, that the density of the exponential distribution is given by
$$
      p_{\lambda}(x) = \begin{cases}\displaystyle
                         \lambda\, {\rm e}^{-\lambda x} & x \ge 0 \\
			0 & x < 0
                      \end{cases} 
$$
for any $\lambda \geq 0$. As a result you should obtain $\kappa_n(X)=\frac{(n-1)!}{\lambda^n}$. 

Hint: Again use the Taylor series for ln and some easy properties of $n!$. Assume $\lambda > t$ in your calculation.

<h3 style="color:rgb(210,90,80)">Calculation (10 points):</h3>

The moment generating function of X, $ M_X(t) $ is:

$ M_X(t) = E \Big[ e^{tX} \Big] = \int_{-\infty}^{\infty} e^{tX} p_{\lambda}(x) dx = \int_{-\infty}^{\infty} e^{tX} \lambda e^{- \lambda X} dx $

then, $ M_X(t) = \int_{-\infty}^{\infty} \lambda e^{(t- \lambda)X} dx = \Big[ \lambda \frac{e^{(t- \lambda)X}}{t- \lambda} \Big]_0^\infty $

So, $ M_X(t) = \begin{cases}\displaystyle
                        \frac{ \lambda}{\lambda - t} & t < \lambda \\
			undefined & t \ge \lambda
                      \end{cases}  $
                      
Therefore, $ M_X^{'}(t) = \frac{d}{dt} \Big( \frac{ \lambda}{ \lambda - t} \Big) = \lambda \frac{d(\lambda - t)^{-1}}{dt} $

$ M_X^{'}(t) = \lambda(-1)(\lambda-t)^{-2}(-1) = \lambda( \lambda - t)^{-2} $

$ M_X^{"}(t) = \lambda(-2)(\lambda-t)^{-3}(-1) = 2\lambda(\lambda-t)^{-3} $

$ M_X^{3}(t) = 2 \lambda (-3)(\lambda-t)^{-4}(-1) = 2(3)\lambda(\lambda-t)^{-4} $

$ M_X^{4}(t) = 2(3)\lambda (-4) (\lambda-t)^{-5}(-1) = (4!)\lambda (\lambda-t)^{-5} $

to be generalized it, 

$ M_X^{n}(t) = (n!) \lambda (\lambda - t)^{-n-1} $ 

Thus, $ \mu_1 = \mu = M_{X}^{'}(0) = \lambda (\lambda)^{-2} = \frac{1}{\lambda} $

$ \mu_2 = M_X^{"}(0) = 2 \lambda(\lambda)^{-3} = \frac{2}{\lambda^2}$

So, to be generalized, 

$ \mu_n = M_X^{n} (0) = (n!)\lambda (\lambda)^{-k-1} = \frac{n!}{\lambda^n} $

<br/>

So, to compute the cumulant generating function

$ ln(M_X(t)) = ln\Big[ \frac{\lambda}{\lambda-t} \Big] $

by expansion using Maclaurin series, $ ln(M_X(t)) = \sum_{n=1}^\infty (n-1)! \Big[\big(\frac{1}{\lambda} \big)^n \Big] \frac{t^n}{n!} $

The cumulant $K_n $ of a random variable X are obtained from the coefficients of $\frac{t^n}{n!} $

Hence, $ K_n = (n-1)! \big( \frac{1}{\lambda} \big)^n  $

Then, the first four cumulants are given below as;

$ K_1 = \frac{1}{\lambda} $

$ K_2 = \big( \frac{1}{\lambda} \big)^2 $

$ K_3 = 2 \big(\frac{1}{\lambda} \big)^3 $

$ K_4 = 6 \big( \frac{1}{\lambda} \big)^4 $ 

To be generalized that, $ K^n(0) = (n-1)! \big (\frac{1}{\lambda} \big)^n $

hence, $ \kappa_n(X) = (n-1)! \big (\frac{1}{\lambda} \big)^n $

Let us mention that the previous argument easily carries over to a Laplace-distributed random variable $X$, which is just a shifted and symmetrized version of the exponential distribution. More preciesly, it has a location parameter $\mu \in \mathbb{R}$ and scaling parameter $\lambda>0$ and is defined by the density 
$$
      p_{\mu,\lambda}(x) = \begin{cases}
                         \frac{\lambda}{2}\, {\rm e}^{-(x-\mu) \lambda} & x \ge \mu \\
			\frac{\lambda}{2}\, {\rm e}^{-(\mu-x)\lambda} & x < \mu.
                      \end{cases} 
$$
By using exactly the same reasoning as before (for each side around $\mu$), one obtains $\kappa_n(X)=\frac{2(n-1)!}{\lambda^n}$, so the cumulants behave the same way as for the exponential distribution.

The next plot should further strengthen your intuition about how deviation from Gaussianity looks like for the Laplace distributed random variables and how this is affected by the parameter $\lambda$. 

Laplace random variables are a typical example of supergaussian random variables, they 
have a spiky density function with heavy tails, i.e. the density function is relatively large at $0$ and at large values, while relatively small for intermediate ones. 

For these kind of distributions (or other kinds of heavy tailed distributions), cumulants are indeed very useful objects to measure the deviation from normality.

However, for not heavy tailed distributions like Gaussians, methods which use entropy as measure of independence might be more suitable.

```python
#nothing to do here
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5,1000)
plt.plot(x, 1/(1 * np.sqrt(2 * np.pi)) *np.exp( - (x )**2 / (2 * 1**2) ), label='Standard normal')
plt.plot(x, 1*np.exp(-abs(x)*1)/2, label='Laplace: lamda=1')
plt.plot(x, 0.5*np.exp(-abs(x)*0.5)/2, label='Laplace: lamda=0.5')

plt.axis([-5, 5, 0, 0.5])
plt.grid()
plt.title('Gauss and Laplace density comparison')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
